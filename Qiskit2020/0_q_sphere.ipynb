```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere
```

```python
# one-qubit system
n = 1 
qc = QuantumCircuit(n)

# Get the statevector from the quantum circuit
statevec = Statevector.from_instruction(qc).data
print(statevec)

# By default, all qubits start in the |0> state = [1, 0].
# That means the first amplitude is 1 (for |0>), 
# the second amplitude is 0 (for |1>).
# This corresponds to 100% probability of measuring 0.
# To get a different state, apply a Hadamard gate.
plot_state_qsphere(statevec)
```

```python
n = 1
qc = QuantumCircuit(n)

# Apply the Hadamard gate to get the equal superposition
# As there is only one qubit, the index is 0.
# So, apply the Hadamard gate on the index 0, which is the first qubit.
qc.h(0)
statevec = Statevector.from_instruction(qc).data
print(statevec)
plot_state_qsphere(statevec)
```

```python
# |00> = [1, 0, 0, 0]^T
# |00>, |01>, |10>, |11>
# |0> (H-> (|0> + |1>)/√2
# qc.h(0) => rightmost ket => apply the Hadamard gate on the right
# (I ⊗ H )|00>   = (|00> + |01>)/√2
n = 2
qc = QuantumCircuit(n)
qc.h(0)
statevec = Statevector.from_instruction(qc).data
print(statevec)
plot_state_qsphere(statevec)
```

```python
# qc.h(0) => leftmost ket => apply the Hadamard gate on the left
# (H ⊗ I )|00>   = (|00> + |10>)/√2
n = 2
qc = QuantumCircuit(n)
qc.h(1)
statevec = Statevector.from_instruction(qc).data
print(statevec)
plot_state_qsphere(statevec)
```

# Bell state

```python
n = 2
qc = QuantumCircuit(n)
qc.h(0)
statevec = Statevector.from_instruction(qc).data
# |00> + |01>/√2
print(f"After applying the Hadarmard Gate: {statevec}")

# apply the CNOT Gate
# control = qubit 0, target = qubit 1
qc.cx(0, 1)
# |00> + |11>/√2
statevec = Statevector.from_instruction(qc).data
print(f"After applying the Hadamard Gate and the CNOT gate: {statevec}")
plot_state_qsphere(statevec)
```

```python
n = 2
qc = QuantumCircuit(n)
qc.h(0)
statevec = Statevector.from_instruction(qc).data
# |00> + |01>/√2
print(f"After applying the Hadamard Gate: {statevec}")

# apply the CNOT Gate
# control = qubit 1, target = qubit 0
qc.cx(1, 0)
statevec = Statevector.from_instruction(qc).data
# |00> + |01>/√2
print(f"After applying the Hadamard Gate and the CNOT Gate: {statevec}")

plot_state_qsphere(statevec)
```

# Bit Flip

```python
n = 2
qc = QuantumCircuit(n)
qc.h(0)
statevec = Statevector.from_instruction(qc).data
# |00> + |01>/√2
print(f"After applying the Hadamard Gate: {statevec}")

# apply the CNOT Gate
# control = qubit 0, target = qubit 1
qc.cx(0, 1)
statevec = Statevector.from_instruction(qc).data
# |00> + |11>/√2
print(f"After applying the Hadamard Gate and the CNOT Gate: {statevec}")

qc.x(1)
statevec = Statevector.from_instruction(qc).data
# Bit flip from |00>+|11>/√2 to |01> + |10>/√2
print(f"After applying the sigma x gate: {statevec}")

plot_state_qsphere(statevec)
```

```python
n = 2
qc = QuantumCircuit(n)
# apply the Hadamard Gate on the leftmost bit
qc.h(0) 
statevec = Statevector.from_instruction(qc).data
# |00> + |01> /√2
print(f"After applying the Hadamard Gate: {statevec}")

# apply the CNOT Gate
# control = qubit 0, target = qubit 1
qc.cx(0, 1)
statevec = Statevector.from_instruction(qc).data
# |00> + |11> /√2
print(f"After applying the Hadamard Gate and the CNOT Gate: {statevec}")

# apply the sigma z Gate
# Phase Flip
qc.z(1)
statevec = Statevector.from_instruction(qc).data
# Phase 180* shift ->> from |00> + |11> /√2 to |00> - |11> /√2
print(f"After applying the Hadamard Gate, the CNOT Gate, and the Sigma Gate: {statevec}")

plot_state_qsphere(statevec)
```

```python
n = 6
qc = QuantumCircuit(n, n)
for i in range(n):
    qc.h(i)
statevec = Statevector.from_instruction(qc).data
plot_state_qsphere(statevec)
```

```python
# Make it more colorful
n = 6
qc = QuantumCircuit(n, n)
for i in range(n):
    qc.h(i)
    qc.rz(2*np.pi/n, i) # 2pi/6, etc
statevec = Statevector.from_instruction(qc).data
plot_state_qsphere(statevec)
```
